#### Exploratory Data Analysis

Kernel Density Estimation (KDE) is a way to estimate the probability density function of a continuous random variable

The 'xlim()' function in pyplot module of matplotlib library is used to get or set the x-limits of the current axes

#### Distribution Plot

```python
fig,axs = plt.subplots(ncols=2,nrows=2,figsize=(10,10))
sns.displot(df['Sales'],color='green', ax=axs[0][0])
sns.displot(df['Profit'],color='red', ax=axs[0][1])
sns.displot(df['Quantity'],color='yellow' ,ax=axs[1][0])
sns.displot(df['Discount'],color='brown',ax=axs[1][1])
axs[0][0].set_title('Sales Distribution' ,fontsize=20)
axs[0][1].set_title('Profit Distribution' ,fontsize=20)
axs[1][0].set_title('Quantity Distribution' ,fontsize=20)
axs[1][1].set_title('Discount Distribution' ,fontsize=20)
plt.show()
```
